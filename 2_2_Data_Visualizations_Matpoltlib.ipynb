# 2.2 Visualizations and MatPlotLib
## 2.2.1 Basic Plot & Scatter
### PLOT

#### MatPlotLib Documentation (it is so pretty!) can be found here
https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
np.random.seed(1221)

df = pd.DataFrame()

df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand']**2
df['rand_shift'] = df['rand'] + 2

df['counts_sq'] = df.index**2
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
plt.plot(df['rand'])
plt.show()
```

np.random.seed(60)

# Make a blank data frame.
df = pd.DataFrame()

# Add a column of random numbers between 0 and 1.
df['rand'] = np.random.rand(10)
df['rand_sq'] = df['rand']**2
df['rand_shift'] = df['rand'] + 2

print(df.head())
# When creating a data frame an index column of counts is created, counting from 0.
# Here we do a few transforms on that index to create some extra columns.
df['counts_sq'] = df.index**2 
df['counts_sqrt'] = np.sqrt(df.index)

```python
plt.plot(df['rand'])
plt.show()
```

```python
plt.plot(df['rand'], color = "purple")
plt.ylim([-.25, 1.1])
plt.ylabel("Values")
plt.title ("Totally Rando Series")

plt.show
```

#### Let's add a second plot!

```python
plt.plot(df['rand'], color = "purple")
plt.plot(df['rand_shift'], color = "orange")
plt.ylim(-.1,3.1)
plt.ylabel("Values")
plt.title("Totally Rando Plot Series")
plt.show()
```

### SCATTER

```python
plt.scatter(x=df["rand"], y=df["rand_sq"])
plt.show()
```

```python
plt.scatter(
    x=df['rand'],
    y=['rand_sq'],
    color = "purple",
    marker = "x", s=10
)
plt.scatter(
    x=df['rand'],
    y=df['rand_shift'],
    color = "green",
    marker = 'x', s=10
)
plt.show()
```

```python
plt.scatter(
    x=df['rand'],
    y=df['rand_sq'],
    color='purple',
    marker='x', s=50
)
plt.scatter(
    x=df['rand'],
    y=df['rand_shift'],
    color='green',
    marker='x', s=50
)
plt.show()
```
